```python
# Import cars data
cars = pd.read_csv('datasets/cars.csv', index_col = 0)

# Print out country column as Pandas Series
print(cars['country'])

# Print out country column as Pandas DataFrame
print(cars[['country']])

# Print out DataFrame with country and drives_right columns
print(cars[['country',"drives_right"]])
```

```python
# Print out first 3 observations
print(cars[0:3])

# Print out fourth, fifth and sixth observation
print(cars[3:6])
```

```python
# Print out drives_right column as Series
print(cars.loc[:,'drives_right'])

# Print out cars_per_cap and drives_right as DataFrame
print(cars.loc[:,["cars_per_cap",'drives_right']])
```

```python
# Create car_maniac: observations that have a cars_per_cap over 500
cpc = cars['cars_per_cap']
many_cars = cars['cars_per_cap'] > 500
car_maniac = cars[many_cars]

# Print car_maniac
print(car_maniac)
```

```python
# Create medium: observations with cars_per_cap between 100 and 500
cpc = cars['cars_per_cap']
between = np.logical_and(cpc >100, cpc <500)
medium = cars[between]

# Print medium
print(medium)
```

```python
#While Loops
# Initialize offset
offset = -6

# Code the while loop
while offset != 0 :
    print("correcting...")
    if offset > 0 :
      offset = offset -1
    else : 
      offset = offset + 1    
    print(offset)
```

```python
#for loops

# areas list
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Change for loop to use enumerate() and update print()
for index,a in enumerate(areas) :
    print("room " + str(index+1) + ":" + str(a))
    
```

```python
#"for loops" over list of lists

# house list
house = [["hallway", 11.25], 
         ["kitchen", 18.0], 
         ["living room", 20.0], 
         ["bedroom", 10.75], 
         ["bathroom", 9.50]]
         
# Build a for loop from scratch
for room, area in house:    
    print("the " + str(room) + " is " + str(area) + " sqm")
```

```python
#looping over dictionaries

# Definition of dictionary
europe = {'spain':'madrid', 'france':'paris', 'germany':'berlin',
          'norway':'oslo', 'italy':'rome', 'poland':'warsaw', 'austria':'vienna' }
          
# Iterate over europe
for country, city in europe.items():    
    print("the capital of " + country + " is " + city)
```

```python
# Import numpy as np
import numpy as np
height = [1,2,3,4,5]
baseball = [["six","five","four","three"],
           [6,5,4,3]]
np_baseball = np.array(baseball)
np_height = np.array(height)

# For loop over np_height
for height in np_height: 
    print(str(height) + " inches")

# For loop over np_baseball
for x in np.nditer(np_baseball): 
    print(x)
```
